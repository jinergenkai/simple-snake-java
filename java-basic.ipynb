### 1. Variables, Data Types, Operators

Basic Variables

```java
int a = 5;
int b = 3;
int abcc$_a = 1;

int sum = a + b;

System.out.println("Sum = " + sum);

var a = "haha";
System.out.print(a);
```

Operators

```java
int x = 10;
x += 5;
System.out.println("x = " + x);
```

### 2. Control Statements

If/Else Example

```java
int score = 70;
if (score >= 50) System.out.println("Pass");
else System.out.println("Fail");
```

Loop Example

```java
for (int i = 1; i <= 5; i--) {
  System.out.println("i = " + i);
}
```

Switch Example

```java
int command = 1;
switch (command) {
  case 1 -> System.out.println("Start");
  case 2 -> System.out.println("Stop");
  default -> System.out.println("Unknown");
}
```

### 3. OOP Basics

Simple Class

```java
class Student {
  String name;
  int age;

  Student(String n, int a) {
    this.name = n;
    this.age = a;
  }

  void introduce() {
    System.out.println("Hi, I'm " + name + " and I'm " + age);
  }
}

Student s1 = new Student("Alice", 20);
s1.introduce();
```

### 4. Exception Handling

Try/Catch

```java
try {
  int value = Integer.parseInt("abc");
} catch (NumberFormatException e) {
  System.out.println("Invalid number!");
}
```

### 5. Collections

List Example

```java
List<String> names = new ArrayList<>();
names.add("Tom");
names.add("Jerry");
System.out.println(names);
```

Set Example

```java
Set<Integer> setNums = new HashSet<>();
setNums.add(1);
setNums.add(1);
setNums.add(2);
System.out.println(setNums);
```

Map Example

```java
Map<String, Integer> mapAges = new HashMap<>();
mapAges.put("Bob", 25);
mapAges.put("Anna", 22);
System.out.println(mapAges);
```

### 6. Streams & Lambdas

Stream Basics

```java
List<Integer> numbers = List.of(1,2,3,4,5);
numbers.stream().filter(n -> n % 2 == 0).forEach(System.out::println);
```
